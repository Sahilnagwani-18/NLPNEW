```python
from automata.fa.dfa import DFA
import string

lowercase = set(string.ascii_lowercase)
all_ascii = set(string.printable)
invalid = all_ascii - lowercase

dfa = DFA(
    states={"q0", "q1", "qE"},
    input_symbols=lowercase.union(invalid),
    transitions={
        "q0": {ch: "q1" if ch in lowercase else "qE" for ch in lowercase.union(invalid)},
        "q1": {ch: "q1" if ch in lowercase else "qE" for ch in lowercase.union(invalid)},
        "qE": {ch: "qE" for ch in lowercase.union(invalid)},
    },
    initial_state="q0",
    final_states={"q1"},
)

print(dfa.accepts_input("cat"))
print(dfa.accepts_input("Dog"))
print(dfa.accepts_input("dog1"))
print(dfa.accepts_input("abc"))
print(dfa.accepts_input(" a"))
```

```python
from automata.fa.nfa import NFA
import string
from graphviz import Digraph

lowercase = set(string.ascii_lowercase)
all_ascii = set(string.printable)
invalid = all_ascii - lowercase

nfa = NFA(
    states={"q0", "q1", "qE"},
    input_symbols=lowercase.union(invalid) | {""},  # Add epsilon symbol
    transitions={
        "q0": {"" : {"q1"},  # epsilon transition
               **{ch: {"qE"} for ch in invalid}},
        "q1": {ch: {"q1"} if ch in lowercase else {"qE"} 
               for ch in lowercase.union(invalid)},
        "qE": {ch: {"qE"} for ch in lowercase.union(invalid)},
    },
    initial_state="q0",
    final_states={"q1"},
)

# Create custom diagram
diagram = Digraph(name="NFA")
diagram.attr(rankdir="LR", fontsize="20")

# Add states
for state in nfa.states:
    if state in nfa.final_states:
        diagram.node(state, shape="doublecircle", fontsize="20")  # Final state
    elif state == "qE":
        diagram.node(state, shape="circle", style="filled", fillcolor="lightcoral", fontsize="20")  # Dead state
    else:
        diagram.node(state, shape="circle", fontsize="20")

# Start arrow
diagram.node("", shape="none")
diagram.edge("", nfa.initial_state)

# Add transitions
for state, paths in nfa.transitions.items():
    for symbol, dest_states in paths.items():
        for dest in dest_states:
            label = "ε" if symbol == "" else symbol
            diagram.edge(state, dest, label=label, fontsize="18")

# Save diagram
diagram.render("nfa_custom", format="png", cleanup=True)
print("Custom NFA diagram saved as nfa_custom.png")
```
