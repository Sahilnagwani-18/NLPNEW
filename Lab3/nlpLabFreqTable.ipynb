```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
```

```python
font_path = r"C:\Windows\Fonts\Nirmala.ttc"
hindi_font = fm.FontProperties(fname=font_path)
```

```python
file_path = r"C:\Sahil\NLP\hindi_tokenized_words2.csv"
df = pd.read_csv(file_path, header=None)
```

```python

words = [w.strip() for w in df[0] if isinstance(w, str) and w != "token"]
```

```python
print(words[:20])
print("Total words:", len(words))
```

```python
freq_dict = {}
for w in words:
    freq_dict[w] = freq_dict.get(w, 0) + 1

sorted_freq = sorted(freq_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
top_100 = sorted_freq[:100]
print("Top 100 words:", top_100)
plt.figure(figsize=(15,6))
plt.bar([w for w,f in top_100], [f for w,f in top_100])
plt.xticks(rotation=90,fontproperties=hindi_font,fontsize=6)
plt.title("Top 100 Most Frequent Words (Before Stopword Removal)")
plt.show()
```

```python
stopWords=[]
```

```python
for w,f in sorted_freq:
    if f>500000:
        stopWords.append(w)
```

```python
print(len(stopWords))
```

```python
removed_stop_words_freq = {}
for w in words:
    if w not in stopWords:
        removed_stop_words_freq[w] = removed_stop_words_freq.get(w, 0) + 1

removed_stop_words_freq = sorted(removed_stop_words_freq.items(), key=lambda x: x[1], reverse=True)
```

```python
top_100 = removed_stop_words_freq[:100]
print("Top 100 words:", top_100)
plt.figure(figsize=(15,6))
plt.bar([w for w,f in top_100], [f for w,f in top_100])
plt.xticks(rotation=90,fontproperties=hindi_font,fontsize=6)
plt.title("Top 100 Most Frequent Words (Before Stopword Removal)")
plt.show()
```
