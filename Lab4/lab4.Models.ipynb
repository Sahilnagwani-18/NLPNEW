```python
import csv

def write_ngrams(input_file, n, output_file):
    with open(input_file, "r", encoding="utf-8") as f_in, \
         open(output_file, "w", encoding="utf-8", newline="") as f_out:

        reader = csv.DictReader(f_in)  
        writer = csv.writer(f_out)
        writer.writerow([f"{n}-gram"])  

        for row in reader:
            sentence = row["reconstructed_sentence"].strip()
            if not sentence:
                continue

            tokens = sentence.split()

            
            tokens = ['<sos>'] * (n - 1) + tokens + ['<eos>']

            for i in range(len(tokens) - n + 1):
                ngram = " ".join(tokens[i:i+n])
                writer.writerow([ngram])
```

```python
from collections import Counter

def load_counts(file_path):
    counts = Counter()
    with open(file_path, "r", encoding="utf-8") as f:
        next(f) 
        for line in f:
            ngram = line.strip()
            counts[ngram] += 1
    return counts

```

```python
V_uni = len(unigram_counts)
V_bi = len(bigram_counts)
V_tri = len(trigram_counts)
V_quad = len(quadgram_counts)


N_uni = sum(unigram_counts.values())
```

```python

def unigram_add_one(token):
    return (unigram_counts.get(token, 0) + 1) / (N_uni + V_uni)

def bigram_add_one(w1, w2):
    return (bigram_counts.get(f"{w1} {w2}", 0) + 1) / (unigram_counts.get(w1, 0) + V_uni)


def trigram_add_one(w1, w2, w3):
    return (trigram_counts.get(f"{w1} {w2} {w3}", 0) + 1) / (bigram_counts.get(f"{w1} {w2}", 0) + V_uni)

def quadgram_add_one(w1, w2, w3, w4):
    return (quadgram_counts.get(f"{w1} {w2} {w3} {w4}", 0) + 1) / (trigram_counts.get(f"{w1} {w2} {w3}", 0) + V_uni)
```

```python

def unigram_add_k(token, k):
    return (unigram_counts.get(token, 0) + k) / (N_uni + k * V_uni)


def bigram_add_k(w1, w2, k):
    return (bigram_counts.get(f"{w1} {w2}", 0) + k) / (unigram_counts.get(w1, 0) + k * V_uni)


def trigram_add_k(w1, w2, w3, k):
    return (trigram_counts.get(f"{w1} {w2} {w3}", 0) + k) / (bigram_counts.get(f"{w1} {w2}", 0) + k * V_uni)


def quadgram_add_k(w1, w2, w3, w4, k):
    return (quadgram_counts.get(f"{w1} {w2} {w3} {w4}", 0) + k) / (trigram_counts.get(f"{w1} {w2} {w3}", 0) + k * V_uni)
```

```python

def unigram_token_type(token):
    token_types = len(unigram_counts)
    return (unigram_counts.get(token, 0) + 1) / (N_uni + token_types)


def bigram_token_type(w1, w2):
    token_types = len(bigram_counts)
    return (bigram_counts.get(f"{w1} {w2}", 0) + 1) / (unigram_counts.get(w1, 0) + token_types)


def trigram_token_type(w1, w2, w3):
    token_types = len(trigram_counts)
    return (trigram_counts.get(f"{w1} {w2} {w3}", 0) + 1) / (bigram_counts.get(f"{w1} {w2}", 0) + token_types)

def quadgram_token_type(w1, w2, w3, w4):
    token_types = len(quadgram_counts)
    return (quadgram_counts.get(f"{w1} {w2} {w3} {w4}", 0) + 1) / (trigram_counts.get(f"{w1} {w2} {w3}", 0) + token_types)
```
